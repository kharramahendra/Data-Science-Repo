```python
def get_outliers(data):
    threshhold = 3 # all the values after 3rd deviation are outliers
    std=np.std(data)
    mean = np.mean(data)
    outliers = []
    for i in data:
        z_score = (i-mean)/std
        if np.abs(z_score)>3:
            outliers.append(i)
    return outliers
```

```python
get_outliers(df['sepal_width']) # we have only one outliers
```
